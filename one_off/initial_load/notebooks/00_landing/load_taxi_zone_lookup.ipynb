```python
import urllib.request 
import shutil 
import os 

# target url to download 
url = 'https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv'

# open connection 
response = urllib.request.urlopen(url)

# crate destnation dir for storing the csv file 

dir_path = "/Volumes/nyctaxi/00_landing/data_sources/lookup"
os.makedirs(dir_path, exist_ok=True)

# defien the full local path {path + file name} where the file will be saved 
local_path =f'{dir_path}/taxi_zone_lookup.csv'

# write the file to the local path 
with open(local_path, 'wb') as f:
    shutil.copyfileobj(response, f)
```
