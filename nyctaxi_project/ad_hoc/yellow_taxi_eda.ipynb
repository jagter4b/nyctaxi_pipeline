```python
from pyspark.sql.functions import *
```

### Which vendor makes the most revenue?


```python
df = spark.read.table("nyctaxi.02_silver.yellow_trips_enriched")

df.groupBy("vendor").agg(round(sum("total_amount"), 2).alias("total_revenue")).orderBy("total_revenue", ascending=False).display()
```

### what is the most popular pickup borough?

```python
df.\
    groupBy("pu_borough").\
    agg(
        count("*").alias("number_of_trips")
    ).\
    orderBy("number_of_trips", ascending=False).\
    display()
```

### what is the most common jounrey (borough to borough)?

```python
df.\
    groupBy("pu_borough", "do_borough").\
    agg(
        count("*").alias("number_of_trips")
    ).\
    orderBy("number_of_trips", ascending=False).\
    display()
```
